```python
import numpy as np
import matplotlib.pyplot as plt

KM = 1e3
distances = np.linspace(1, 200*KM, 500)



fig = plt.figure()
ax = fig.add_subplot(111)

for exp in (1, 2):
    ax.plot(distances/KM, 1/(distances**exp), label=f"exp={exp}")

ax.legend()
ax.grid(alpha=0.3)
ax.set_xlabel("Distance [km]")
ax.set_ylabel("Weight")
```
